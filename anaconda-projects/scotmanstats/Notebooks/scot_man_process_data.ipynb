# Code setup

```python
import pandas as pd
import numpy as np
```

# Load and clean data

```python
scotman_rev_df = pd.read_csv('data/ScottishManufRevInScotland_bands_3.csv')
cols = scotman_rev_df.columns.drop('Description')
scotman_rev_df[cols] = scotman_rev_df[cols].apply(pd.to_numeric, errors='coerce').fillna(0)
scotman_rev_df.index = scotman_rev_df['Description']
#scotman_df.drop(['Year', 'Division/\nSection', 'Description'], axis=1, inplace=True)
scotman_rev_df.drop(['Unnamed: 11','Description'], axis=1, inplace=True)
cols = scotman_rev_df.columns
cols_fixed = [col.replace('\n', ' ').replace('�','£') for col in cols]
cols_fixed
scotman_rev_df.columns = cols_fixed
#scotman_rev_df

scotman_emp_df = pd.read_csv('data/ScottishManufEmpInScotland.csv')
cols = scotman_emp_df.columns.drop('Description')
scotman_emp_df[cols] = scotman_emp_df[cols].apply(pd.to_numeric, errors='coerce').fillna(0)
scotman_emp_df.index = scotman_emp_df['Description']
scotman_emp_df.drop(['Year', 'Division/\nSection', 'Unnamed: 13', 'Description'], axis=1, inplace=True)

scotman_df = scotman_rev_df.join(scotman_emp_df)
scotman_df
```

# Fill in Total Revenues in bands

```python
scotman_totals_df = pd.read_csv('data/ScottishManufInScotlandTotals_clean2.csv')
cols = scotman_totals_df.columns.drop('Description')
scotman_totals_df[cols] = scotman_totals_df[cols].apply(pd.to_numeric, errors='coerce').fillna(0)
scotman_totals_df.index = scotman_totals_df['Description']
cols = scotman_totals_df.columns
cols_fixed = [col.replace('\n', ' ').replace('�','£') for col in cols]
scotman_totals_df.columns = cols_fixed
scotman_totals_df['Total Scottish turnover (£)'] = 1000000 * scotman_totals_df['Total Scottish turnover (£m)']
scotman_totals_df.drop(['Description', 'Total Scottish turnover (£m)'], axis=1, inplace=True)
scotman_totals_df
```

```python
cols = scotman_df.columns
cols
```

```python
centres = [25000, 75000, 150000, 600000, 3000000, 7500000, 15000000, 22200000, 37500000, 0, 0, 3.5, 7.4, 14.5, 34.5, 74.5, 149.5, 224.5, 374.5, 0]
centres_dict = dict(zip(cols, centres))
centres_dict
```

```python
def estimate_total_revs(row):
    total = sum([row[col] for col in row.index if 'Estimated Revenues' in col])
    return total

def estimate_total_emps(row):
    total = sum([row[col] for col in row.index if 'Estimated Employees' in col])
    return total

def estimate(row):
    for col in row.index:
        if col in scotman_rev_df.columns:
            new_col = col.replace('Number', 'Estimated Revenues')
        else:
            new_col = col.replace('Number', 'Estimated Employees')
        row[new_col] = row[col] * centres_dict[col]
    return row
```

```python
scotman_df = scotman_df.apply(estimate, axis=1)
scotman_df
```

```python
scotman_df.columns
```

```python
scotman_df['temp estimated total revenues'] = scotman_df.apply(estimate_total_revs, axis=1)
scotman_df['temp estimated total employees'] = scotman_df.apply(estimate_total_emps, axis=1)
scotman_df
```

```python
scotman_df['Estimated Revenues: £50,000,001+'] = scotman_totals_df['Total Scottish turnover (£)'] - scotman_df['temp estimated total revenues']
scotman_df['Total Revenues'] = scotman_totals_df['Total Scottish turnover (£)']

scotman_df['Estimated Employees: 500+ employees'] = scotman_totals_df['Total Scottish employment'] - scotman_df['temp estimated total employees']
scotman_df['Total Employees'] = scotman_totals_df['Total Scottish employment']

scotman_df.drop(['temp estimated total revenues', 'temp estimated total employees'], axis=1, inplace=True)

scotman_df
```
